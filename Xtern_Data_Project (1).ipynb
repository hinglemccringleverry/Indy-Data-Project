```python
import ipyleaflet
#We will be using the ipyleaflet package for this project
```

```python
from ipyleaflet import Map, Marker, AwesomeIcon, AntPath
#Import the map, marker, and awesomeicon package to place categorical markers
#Import antpath to visualise the routes
```

```python
indy = Map(center=(39.770696092192416, -86.15679451871569),zoom=12)
```

```python
mexican = AwesomeIcon(
    marker_color='red',
    spin=False
)
american = AwesomeIcon(
    marker_color='blue',
    spin=False
)
hotdog = AwesomeIcon(
    marker_color='orange',
    spin=False
)
other = AwesomeIcon(
    marker_color='green',
    spin=False
)
#Identifies the cuisine variables as 4 types: Mexican, American, Hotdog, and "Other". 
#Places were split into these variable types based on if there was more than 1 unique type of cuisine for the resturaunt dataset
#"Other" cuisine types are the combination of unique cuisines that had less than 2 entries
#The goal should be to hit one resturaunt of these categories for each day
```

```python
clustertruck = Marker(icon=american,location=(39.78078759439361, -86.1563859064785))
nightowl = Marker(icon=american,location=(39.761859366166384, -86.1578122858574))
taq_mi_costa = Marker(icon=mexican,location=(39.767795371281316, -86.18644284167945))
taq_el_taco = Marker(icon=mexican,location=(39.76532662872753, -86.20491195757374))
dogs_to_go = Marker(icon=hotdog,location=(39.78062875793833, -86.2160718472253))
fire_express = Marker(icon=other,location=(39.78046349996953, -86.14351505702115))
grub_house = Marker(icon=other,location=(39.793883942047295, -86.14419128030991))
mr_and_ms = Marker(icon=hotdog,location=(39.77100805971355, -86.16961087828182))
greiner = Marker(icon=other,location=(39.7389123231328, -86.13972468031194))
salvador = Marker(icon=other,location=(39.81215520230169, -86.2196840082075))
tacos_y = Marker(icon=mexican,location=(39.75139274659921, -86.19762503373296))
los_dos = Marker(icon=mexican,location=(39.76066441174097, -86.22451668400865))
#Importing the coordinates of each location from Google Maps
```

```python
indy.add_layer(clustertruck)
indy.add_layer(nightowl)
indy.add_layer(taq_mi_costa)
indy.add_layer(taq_el_taco)
indy.add_layer(dogs_to_go)
indy.add_layer(fire_express)
indy.add_layer(grub_house)
indy.add_layer(mr_and_ms)
indy.add_layer(greiner)
indy.add_layer(salvador)
indy.add_layer(tacos_y)
indy.add_layer(los_dos)
#Import every location as a marker on the map
```

```python
sunday = ant_path = AntPath(
    locations=[[39.76532662872753, -86.20491195757374],[39.78062875793833, -86.2160718472253],[39.81215520230169, -86.2196840082075]],
            dash_array=[1, 10],
        delay=1000,
        color='#7590ba',
        pulse_color='#3f6fba'
              )
indy.add_layer(sunday)
#Sunday's path with the animation giving the direction/sequence
```

```python
saturday = ant_path = AntPath(
    locations=[[39.767795371281316, -86.18644284167945],[39.78078759439361, -86.1563859064785],[39.78046349996953, -86.14351505702115]],
            dash_array=[1, 10],
        delay=1000,
        color='red',
        pulse_color='black',
        reverse = True
              )
indy.add_layer(saturday)
#Saturdays's path with the animation giving the direction/sequence
```
